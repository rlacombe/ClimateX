```python
import pandas as pd

```

```python
cleaned_test_set_with_context = pd.read_csv("data/ipcc_cleaned_test_set_with_context.csv")
original_extracted_sentences = pd.read_csv("data/text_processing/sentences_with_ratings_all.csv")

```

```python
cleaned_test_set_with_context.head()
```

```python
original_extracted_sentences.head()
```

```python
final_test_set = cleaned_test_set_with_context[["statement_idx", "report", "page_num", "sent_num", "confidence", "score", "split", "context", "statement", "rewrite"]]
original_extracted_sentences.reset_index(inplace=True)
original_extracted_sentences.rename(columns={"index": "statement_idx"}, inplace=True)
final_test_set = final_test_set.merge(original_extracted_sentences[["statement_idx", "text"]], on="statement_idx")

final_test_set.rename(columns={"text": "original_statement"}, inplace=True)
final_test_set['final_statement'] = final_test_set.apply(lambda row: row['rewrite'] if pd.notna(row['rewrite']) else row['statement'], axis=1)
final_test_set = final_test_set[["statement_idx", "report", "page_num", "sent_num", "original_statement", "final_statement", "confidence", "score", "split", "context"]]


```
